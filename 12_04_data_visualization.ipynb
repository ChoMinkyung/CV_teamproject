```python
# 총 5가지 평가지표, 4개 모델 비교
# importing package
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 1, figsize=(18,13))
# create data
x = np.arange(5)
ML_value = [0.7414, 0.7421, 0.7410, 0.7410, 0.8274]
DNN_value = [0.6731, 0.6722, 0.6631, 0.6720, 0.7820]
Transformer_value = [0.7422, 0.7474, 0.7426, 0.7435, 0.8283]
Ensemble_value = [0.7516, 0.7540, 0.7510, 0.7516, 0.8341]
width = 0.2

ML_value = np.dot(ML_value, 100)
DNN_value = np.dot(DNN_value, 100)
Transformer_value = np.dot(Transformer_value, 100)
Ensemble_value = np.dot(Ensemble_value, 100)

  
# plot data in grouped manner of bar type
plt.bar(x-0.3, ML_value, width, color='lightsteelblue')
plt.bar(x-0.1, DNN_value, width, color='cornflowerblue')
plt.bar(x+0.1, Transformer_value, width, color='royalblue')
plt.bar(x+0.3, Ensemble_value, width, color='midnightblue')
print(ax.containers[0])
plt.bar_label(ax.containers[0], size=12, fmt='%.2f')
plt.bar_label(ax.containers[1], size=12, fmt='%.2f')
plt.bar_label(ax.containers[2], size=12, fmt='%.2f')
plt.bar_label(ax.containers[3], size=12, fmt='%.2f', color='red')
plt.ylim(60, 90)
plt.xticks(x, ['Accuracy', 'Precision', 'Recall', 'F1 score', 'AUC'], fontsize=16)
plt.xlabel("Metrics", fontsize=17)
plt.ylabel("Score (%)", fontsize=17)
plt.legend(["SVM", "CNN+Bi-LSTM", "KoBERT", "Ensemble"], fontsize=16)
plt.title('Performance Comparison', size=20)

plt.show()
```
